```python
def metrics(text, genre):
    #number of distinct words in a lyric
    counter = Counter(text[text['genre'] == genre]['lyrics'])
    distinct_words = len(counter)

    #average number of tokens in a lyric
    len_toks = sum(text[text['genre'] == genre]['t-lyric'].str.len())/len(text[text['genre']== genre]['t-lyric'])

#CODE BELOW INCOMPLETE    
    
    #average word length 
    words = text.split()
    avg_word_len = sum(len(word) for word in words) / len(words)
    
    #average line length
    #get the number of occurrences of each length of line
    text[text['genre'] == genre]['lyrics'].str.split(pat = "/")
    count = text.str.split().apply(len).value_counts()
    #find the average length of line
    avg_line_len = sum(count * count.index)/len(text)
    
    return distinct_words, len_toks, distinct_toks, avg_word_len
```
